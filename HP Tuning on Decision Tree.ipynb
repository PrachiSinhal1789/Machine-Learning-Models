```python
import pandas as pd
import numpy as np
from sklearn.model_selection import GridSearchCV,cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,classification_report
```

```python
df=pd.read_excel('KTMData.xlsx')
```

```python
df=pd.get_dummies(df,columns=['Gender','Occupation','Phone Type','Current Bike','Relationship'])
```

```python
X=df.drop(['Response'],axis=1)
```

```python
y=df['Response']
```

```python
param={'criterion':['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],'max_depth':[1,3,5,7,10,12,14,16]}
```

```python
model=GridSearchCV(DecisionTreeClassifier(),param)
```
